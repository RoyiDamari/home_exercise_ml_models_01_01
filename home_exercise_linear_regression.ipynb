```python
# Exercise 1 - B

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Our data
print("Solution B")
advertising_investment = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50]).reshape(-1, 1)  # Advertising Investment
sales_growth = np.array([25, 30, 40, 45, 50, 60, 65, 70, 80])  # Sales Growth

# Create regression model
model = LinearRegression()
model.fit(advertising_investment, sales_growth)

# Print results
print(f"Slope (m): {model.coef_[0]:.2f}")
print(f"Intercept (b): {model.intercept_:.2f}")

# Calculate equation
equation = f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}"
print(f"Line equation: {equation}")

x = 60
predicted_sales_grows = model.coef_[0] * x + model.intercept_
print(f"Investment of 60 in advertising will produce sales growth of {predicted_sales_grows:.2f}")

# Predict hours needed to get score of 100
sales_to_predict = 100
advertising_needed = (sales_to_predict - model.intercept_) / model.coef_[0]
print(f"To get a sales growth of 100, approximately {advertising_needed:.2f} of investment in advertising are needed")
print()

# Create the graph
plt.figure(figsize=(10, 10))
plt.scatter(advertising_investment, sales_growth, color='blue', label='Data points')
plt.plot(advertising_investment, model.predict(advertising_investment), color='red', label='Regression line')

# Add prediction point
plt.scatter([[advertising_needed]], [100], color='green', s=100, label='Our prediction')

# Add labels in English
plt.title('Linear Regression - Advertising Investment (1000 ILS) vs. Sales Growth (1000 ILS)')
plt.xlabel('Advertising Investment (1000 ILS)')
plt.ylabel('Sales Growth (1000 ILS)')
plt.grid(True)
plt.legend()

# Display equation on the graph
plt.text(10, 95, equation, fontsize=12)

plt.show()
print()

# D

print("Solution D")
advertising_investment = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50])
coefficients = np.polyfit(advertising_investment, sales_growth, deg=1)

slope = coefficients[0]
intercept = coefficients[1]

print(f"Slope (b1): {slope:.2f}")
print(f"Intercept (b0): {intercept:.2f}")
```

```python
# exercise 2 - A

import pandas as pd
import scipy
import seaborn as sns

# Our data
print("Solution A:")
phone_call_hours = np.array([2, 3, 4, 5, 6]).reshape(-1, 1)  # Phone Call Hours
profit = np.array([50, 70, 90, 100, 110])  # Profit Achieved (K$)

# Create regression model
model = LinearRegression()
model.fit(phone_call_hours, profit)

# Print results
print(f"Slope (m): {model.coef_[0]:.2f}")
print(f"Intercept (b): {model.intercept_:.2f}")

# Calculate equation
equation = f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}"
print(f"Line equation: {equation}")
print()

# B
print("Solution B:")
x = 8
predicted_profit = model.coef_[0] * x + model.intercept_
print(f"Phone cal of 8 hours will produce profit of {predicted_profit:.2f}")
print()

# C
print("Solution C:")
# Predict hours needed to get score of 100
profit_to_predict = 150
phone_call_hours_needed = (profit_to_predict - model.intercept_) / model.coef_[0]
print(f"To get a profit of 150k, approximately {phone_call_hours_needed:.2f} hours of phone call are needed")
print()

# D
print("Solution D:")
# Create the graph
plt.figure(figsize=(10, 10))
plt.scatter(phone_call_hours, profit, color='blue', label='Data points')
plt.plot(phone_call_hours, model.predict(phone_call_hours), color='red', label='Regression line')

# Add prediction point
plt.scatter([[phone_call_hours_needed]], [150], color='green', s=100, label='Our prediction')

# Add labels in English
plt.title('Linear Regression - Phone Call Hours vs. Profit (K$)')
plt.xlabel('Phone Call Hours')
plt.ylabel('Profit (K$)')
plt.grid(True)
plt.legend()

# Display equation on the graph
plt.text(2, 135, equation, fontsize=12)

plt.show()
print()

# E
print("Solution E:")
phone_call_hours = np.linspace(6, 20, 15)  # Phone Call Hours
# profit = np.array([50, 70, 90, 100, 110])  # Profit Achieved (K$)

def predict_profit(a: float, b: float, hours: list[float]) -> list[float]:
    list_of_profit: list = []
    for hour in hours:
        y = a * hour + b
        list_of_profit.append(y)
    return list_of_profit

profit = predict_profit(model.coef_[0], model.intercept_, phone_call_hours)

df = pd.DataFrame({'phone_call_hours': phone_call_hours, 'profit': profit})

#create regplot
p = sns.regplot(data=df, x=df.phone_call_hours, y=df.profit)

#calculate slope and intercept of regression equation
slope, intercept, r, p, sterr = scipy.stats.linregress(x=p.get_lines()[0].get_xdata(),
                                                       y=p.get_lines()[0].get_ydata())

#add regression equation to plot
plt.text(6, 295, 'y = ' + str(round(slope,3)) +  'x' ' + ' + str(round(intercept,3)))
```
